```python
%matplotlib inline 
#graficos en el notebook

import triangle #pip install triangle
import triangle.plot as plot
import matplotlib.pyplot as plt
```

```python
data = {'vertices':[[0, 0], [0, 1], [1, 1], [1, 0]],
        'segments':[[0,1],[1,2],[2,3],[3,0]]}
```

```python
#triangulacion de delaunay de un conjunto de puntos
tri = triangle.delaunay(data['vertices'])
```

```python
# [1,0,3] significa el triÃ¡ngulo [0, 1]->[0, 0]->[1, 0]->[0, 1]
tri
```

```python
#triangulacion de delaunay con restricciones
a = 0.001
tri = triangle.triangulate(data)
tri
```

```python
#triangulacion de delaunay con restricciones de volumen maximo 0.2
a = 0.001
tri = triangle.triangulate(data, opts='a.2')
plot.plot(plt.axes(), **tri)
```

```python
#triangulacion de delaunay con restricciones de volumen maximo 0.02
a = 0.001
tri = triangle.triangulate(data, opts='a.02')
plot.plot(plt.axes(), **tri)
```
